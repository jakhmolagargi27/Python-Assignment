```python
df[categorical_columns] = df[categorical_columns].fillna('Unknown')
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
avg_price_year = df.groupby('year')['sellingprice'].mean()
avg_price_year.plot(kind='bar', figsize=(10,5))
plt.title("Average Selling Price by Year")
plt.xlabel("Year")
plt.ylabel("Average Selling Price")
plt.show()

#Insight: Newer cars generally have higher selling prices
```

```python
df['odometer_bin'] = pd.cut(df['odometer'], bins=10)
avg_price_odometer = df.groupby('odometer_bin')['sellingprice'].mean()
avg_price_odometer.plot(kind='line', figsize=(10,5))
plt.title("Average Selling Price by Odometer")
plt.xlabel("Odometer Range")
plt.ylabel("Average Selling Price")
plt.show()

#Insight: As odometer increases, selling price decreases
```

```python
state_counts = df['state'].value_counts()
state_counts.plot(kind='bar', figsize=(10,5))
plt.title("Number of Cars Sold by State")
plt.xlabel("State")
plt.ylabel("Number of Cars")
plt.show()
state_counts.head(3)
```

```python
df['condition_range_5'] = pd.cut(df['condition'], bins=range(0, 55, 5))
avg_price_condition = df.groupby('condition_range_5', observed=True)['sellingprice'].mean()
avg_price_condition.plot(kind='bar', figsize=(10,5))
plt.title("Average Selling Price by Condition Range (Size 5)")
plt.xlabel("Condition Range")
plt.ylabel("Average Selling Price")
plt.show()

#Insight: Better condition cars fetch higher prices
```

```python
df['condition_range_10'] = pd.cut(df['condition'], bins=range(0, 60, 10))
count_condition = df['condition_range_10'].value_counts().sort_index()
count_condition.plot(kind='bar', figsize=(10,5))
plt.title("Number of Cars Sold by Condition Range (Size 10)")
plt.xlabel("Condition Range")
plt.ylabel("Number of Cars")
plt.show()

#Insight: Most cars fall in mid-to-high condition ranges
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='color', y='sellingprice', data=df)
plt.title("Selling Price Distribution by Car Color")
plt.xticks(rotation=45)
plt.show()

#Insight: Some colors show wider price variation
```

```python
Q1 = df['sellingprice'].quantile(0.25)
Q3 = df['sellingprice'].quantile(0.75)
IQR = Q3 - Q1
df_no_outliers = df[(df['sellingprice'] >= Q1 - 1.5*IQR) & (df['sellingprice'] <= Q3 + 1.5*IQR)]
plt.figure(figsize=(10,5))
sns.boxplot(x='color', y='sellingprice', data=df_no_outliers)
plt.title("Selling Price by Color (Outliers Removed)")
plt.xticks(rotation=45)
plt.show()
```
